```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-1.5, 1.5, 301) # x and y span
X1, X2 = np.meshgrid(a,a) # coordinate matrices
dx1 = X1 - X1**2 # x1 dot
dx2 = X1 + X2 # x2 dot
fig = plt.figure()
ax = fig.add_subplot()
ax.streamplot(X1, X2, dx1, dx2, density= 2)
ax.grid('on')
ax.plot([0], [0], 'ro',)
ax.annotate('fixed point', xy=(0, 0), xytext=(.25, .25),
            arrowprops=dict(facecolor='black', shrink=0.03, width=0.7, headwidth=4, headlength=4))
ax.plot([1], [-1], 'ro',)
ax.annotate('fixed point', xy=(1, -1), xytext=(.45, -.75),
            arrowprops=dict(facecolor='black', shrink=0.03, width=0.7, headwidth=4, headlength=4))
```
