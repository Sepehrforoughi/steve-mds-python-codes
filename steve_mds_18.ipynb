```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-5,5,101) # initial conditions span
x0 , y0 = np.meshgrid(a, a) # coordinate matrices
t = .1 
u = np.exp(t)*x0 # evaluate xdot at time t = 0.1s
v = -np.exp(-t)*y0
plt.streamplot(x0, y0, u, v, linewidth=0.5, density=2)
plt.grid('on')
```

```python
a = np.linspace(-3,3,30)
x , y = np.meshgrid(a, a)
t = .1
u = np.exp(t)*x
v = -np.exp(-t)*y
plt.quiver(x, y, u, v)
plt.grid('on')
```
