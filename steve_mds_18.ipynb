```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.linspace(-5,5,101) # x and y span
X , Y = np.meshgrid(a, a) # coordinate matrices
t = .1 
u = np.exp(t)*X # evaluate xdot at time t = 0.1s
v = -np.exp(-t)*Y
fig = plt.figure()
ax = fig.add_subplot()
ax.streamplot(X, Y, u, v, density=2)
ax.grid('on')
ax.plot([0], [0], 'ro',)
ax.annotate('fixed point', xy=(0, 0), xytext=(0.5, 0.75),
            arrowprops=dict(facecolor='black', shrink=0.03, width=0.7, headwidth=4, headlength=4))
```
